# <a id='toc1_'></a>[Как компилируется программа на C++?](#toc0_)

## <a id='toc1_1_'></a>[Этап No 1: препроцессор](#toc0_)

∙Язык препроцессора – это специальный язык программирования, встроенный в C++.  
∙Препроцессор работает с кодом на C++ как с текстом.  
∙Команды языка препроцессора называют директивами, все директивы начинаются со знака #.  
∙Директива #include позволяет подключать заголовочные файлы к файлам кода.  

    1.#include <foo.h> — библиотечный заголовочный файл (библиотечный файл берётся из "библиотечной папки"),
    2.#include "bar.h"— локальный заголовочный файл (путь берется относительно исходного .cpp).
∙Препроцессор заменяет директиву #include "bar.h" на содержимое файла bar.h.


## <a id='toc1_2_'></a>[Этап 2: компиляция](#toc0_)

∙На вход компилятору поступает код на C++ после обработки препроцессором.  
∙Каждый файл с кодом компилируется отдельно инезависимо от других файлов с кодом.  
∙Компилируется только файлы с кодом (т.е.*.cpp).  
∙Заголовочные файлы сами по себе ни во что не компилируются, только в составе файлов с кодом.  
∙На выходе компилятора из каждого файла с кодом получается “объектный файл” — бинарный файл со скомпилированным кодом (с расширением .o или .obj).

\*) компилятор будет обрабатывать только *.cpp файлы, в которые препроцессор подставит *.hpp файлы. По каждому cpp сделают obj файл. А линковщик потом всё объединит в один исполняемый файл.

\*\*) функции в объектном файле представлены по именам. Любая функция, которая передаётся компилятору, компилируется отдельно. Потом уже после линковки остаются только те, которые используются. Из одинаковых инлайн-функции выберется любая.

## <a id='toc1_3_'></a>[Этап 3: линковка (компоновка)](#toc0_)

∙На этом этапе все объектные файлы объединяются в один исполняемый (или библиотечный) файл.  
∙При этом происходит подстановка адресов функций в места их вызова

    .void  foo() {
        bar ();     // в соответствующем .obj тут будет адрес из второго .obj (call 0x0baf27)
    }
    
    // другой файл - другой .obj
    void  bar() {   }

∙По каждому объектному файлу строится таблица всех функций, которые в нём определены.    
∙На этапе компоновки важно, что каждая функция имеет уникальное имя (в чистом С такой проблемы нет - одинаковые имена не могут использоваться).
∙В C++ может быть две функции с одним именем, но разными параметрами.  
∙Имена функций искажаются (mangle) таким образом, что в их имени кодируются их параметры.  
    Например, компилятор GCC превратит имя функции void foo(int, double) {} в _Z3fooid.  
∙Аналогично функциям в линковке нуждаются глобальные переменные.  

∙Точка входа— функция, вызываемая при запуске программы. По умолчанию — это функция main. Допустимы только 2 варианта:

    int main()
    {
        return  0;
    }
    
    или
    
    int main(int argc , char ** argv)
    {
        return  0;
    }


## <a id='toc1_4_'></a>[Команды g++ (набор инструментов из пакета gcc, включает препроцессор, компилятор, линковщик и др.):](#toc0_)

    $ g++ -E name.cpp -o name_preproc.cpp                   # только препроцессор
    $ g++ -c name.cpp                                       # компилятор (для каждого cpp-файла) - выход name.o

    $ g++ name1.cpp name2.cpp ... main.cpp -o programm      # компиляция в объектные файлы и их линковка в исполняемый файл

## <a id='toc2_5_'></a>[Преобразование встроенных типов в операторах](#toc0_)

    int a = 20;
    int b = 50;
    double d = a / b;           // d = 0, оба аргумента целочисленные, а значит деление целочисленное 
    double d = (double)a / b;   // d = 0.4

Операторы для встроенных типов C++ всегда работают с одинаковыми типами аргументов. Если аргументы имеют разные типы, то происходит преобразование типов (promotion).  

### <a id='toc2_5_1_'></a>[Правило преобразования встроенных типов в операторах](#toc0_)

Рассмотрим выражение (a + b), где вместо '+' может стоять любой другой подходящий оператор.

    Если один из аргументов имеет числовой тип с плавающей точкой, то второй аргумент приводится к этому типу (например, при сложении double и int значение типа int приводится к double).
    Если оба аргумента имеют числовой тип с плавающей точкой, то выбирается наибольший из этих типов (например, при сложении double и float значение типа float приводится к double).
    Если оба аргумента целочисленные, но их типы меньше int, то оба аргумента приводятся к типу int (например, при сложении двух значений типа char они оба сначала приводятся к int).
    Если оба аргумента целочисленные, то аргумент с меньшим типом приводится к типу второго аргумента (например, при сложении long и int значение типа int приводится к long).
    Если оба аргумента целочисленные и имеют тип одного размера, то предпочтение отдаётся беззнаковому типу (например, при сложении int и unsigned int значение типа int приводится к unsigned int).

### <a id='toc2_5_2_'></a>[Несколько важных следствий](#toc0_)

    Следите за тем, какие типы участвуют в выражении, от этого может зависеть его значение.
    Не стоит использовать целочисленные типы меньше int в арифметических выражениях, они всё равно будут приведены к int.
    Не стоит смешивать unsigned и signed типы в одном выражении, это может привести к неприятным последствиям.

Пример

    unsigned from = 100;
    unsigned to = 0;
    for (int i = from; i >= to; --i) {  ....  }

\*) цикл — бесконечный: в условии цикла проверяется i >= to, где тип i — int, а тип to — unsigned int. Так как операторы всегда применяются к одинаковым встроенным типам, то в данном случае значение переменной i (в соответствии с правилом 5) будет преобразовано к unsigned int, а значения этого типа всегда неотрицательны, т.е. >= 0. Другими словами, когда пременная i станет равной -1, то в условии будет проверяться (unsigned)-1 >= 0, где  (unsigned)-1 = UINT_MAX (UINT_MAX — максимальное значение, которое может принимать переменная типа unsigned int).

## <a id='toc2_8_'></a>[Циклы](#toc0_)

    int  squares = 0;

    while (true) {
        ...
    }
    for (;;) {
        ...
    }

Ещё один оператор, который можно использовать с циклами — это оператор continue. Оператор continue прерывает текущую итерацию цикла и переходит к следующей.   
Например, можно посчитать сумму всех чисел от 1 до 100, которые не делятся на 17 или 19, следующим образом.  

    int sum = 0;
    for ( int i = 1; i <= 100; ++i ) {
        if ( (i % 17 == 0) || (i % 19 == 0) )
           continue; // перейти к следующей итерации
        sum += i;
    }
    
### <a id='toc2_8_1_'></a>[Цикл do-while](#toc0_)

В С++ существует вариация цикла while, которая называется do-while. В отличие от обычного while в do-while условие проверяется не до, а после итерации. Т.е. такой цикл всегда имеет хотя бы одну итерацию.  

    int i = 10;
    int sum = 0;
    do {
        sum += i;
    } while(i < 10);
    // sum = 10
\*) Иногда цикл do-while используется в макросах (см. следующий урок) для того, чтобы объединить несколько инструкций в одну. Действительно, если мы обернём инструкции в do-while c ложным условием, то это будет удобной синтаксической конструкцией, все инструкции которой выполнятся ровно один раз.

    do { /* code */ } while(false); // тело цикла выполнится ровно один раз

## <a id='toc2_10_'></a>[Посимвольный ввод](#toc0_)

Оператор чтения из потока ввода (>>), обычно, игнорирует пробельные символы (это поведение можно изменить с помощью std::noskipws). Иногда это нежелательно. Для организации посимвольного ввода можно использовать метод get потока ввода так, как это показано в коде ниже

    char c = '\0';              // конец ввода (строки)
    while (cin.get(c)) {        // на каждой итерации считываем один символ в переменную c
        /* здесь можно пользоваться значением прочитанным в переменную c */
        if (c != 'a')
            cout << c;          // выводим символ, если он не равен 'a'
    }

Метод get принимает на вход ссылку на переменную типа char.   
Значение, которое возвращает метод get может быть интерпретировано как логическое значение. Это значение интерпретируется как false, если при чтении произошла ошибка (редкий случай) или достигнут конец потока ввода.  
Программа достигнет конца потока ввода, когда прочитает весь файл целиком. Если же программа читает ввод с клавиатуры, то конец потока ввода определяется специальным нажатием клавиш, например, в Linux это, как правило, сочетание Ctrl+D, а в Windows — Ctrl-Z.




## <a id='toc2_11_'></a>[Функции](#toc0_)

∙В сигнатуре функции указывается тип возвращаемого значения и типы параметров.  
∙Ключевое слово return возвращает значение.
    
    double square(double x) {
        return x * x;
    }

∙Переменные, определённые внутри функций — локальные.  
∙Функция может возвращать void (процедура).  
∙Параметры передаются по значению (копируются).  
    
    void  strange(double x, double y) {
        x = y;
    }

Функции иногда используются для того, чтобы реализовать выход из вложенных циклов. Оператор return позволяет немедленно завершить выполнение функции в любом её месте, в том числе внутри вложенных циклов. 

Предположим, что мы хотим проверить, есть ли такие целые i,j>0 : что i\**2+j\**2=x.  Мы можем сделать это просто перебрав все возможные значения для пар (i,j). 

    bool found = 0;
    for (int i = 1; i*i < x; ++i) 
    {
        for (int j = i; j*j < x; ++j) 
        {
             if ( i*i + j*j == x )
             {
                 found = true;
                 break; // выходим из вложенного цикла
             }
        }
        if (found) // выходим из внешнего цикла
             break;
    }

Если обернуть этот код в функцию, то вместо оператора break можно будет использовать оператор return. Не нужно делать выход из каждого цикла по отдельности — как только мы найдём подходящие i и j, мы сразу завершаем выполнение функции и выходим изо всех циклов.

    bool check(int x)
    {
        for (int i = 1; i*i < x; ++i) 
        {
            for (int j = i; j*j < x; ++j) 
            {
                if ( i*i + j*j == x )
                    return true;           
            }
        }
        return false;
    }

## <a id='toc2_12_'></a>[Макросы](#toc0_)

∙Макросами в C++ называют инструкции препроцессора.  
∙Препроцессор C++ является самостоятельным языком, работающим с произвольными строками.  
∙Макросы можно использовать для определения "функций":  

    int  max1(int x, int y) {
        return x > y ? x : y;}
    
    #define  max2(x, y)    x > y ? x : y    // разбирает препроцессор (он не знает про типы, считает их строками)
    
    a = b + max2(c, d);       // b + c > d ? c : d; поэтому он просто подставит макрос, что нарушит ожидаемый порядок

∙Препроцессор “не знает” про синтаксис C++
∙Параметры макросов нужно оборачивать в скобки:
    
    #define  max3(x, y)    ((x) > (y) ? (x) : (y))
    
∙Это не избавляет от всех проблем:
    
    int a = 1;
    int b = 1;
    int c = max3 (++a, b);  // ожидается с=2
    // c = ((++a) > (b) ? (++a) : (b))  по факту инкремент подставится дважды

∙Определять функции через макросы — плохая идея. Еще пример
    
    #define sqr(x) x * x
    sqr(3 + 0) // 3 + 0 * 3 + 0 = 3 
∙Макросы можно использовать для условной компиляции:
    
    #ifdef  DEBUG
        // дополнительные проверки
    #endif

Главный минус макроса в том, что он не является "функцией", а именно часть кода, которая будет вставляться везде дальше, где будет он использован. За счет этого код будет разрастаться в размерах на пустом месте.

А плюс его заключается в его скорости исполнения, поскольку по сравнению с функцией, не будет тратиться время на его вызов.
